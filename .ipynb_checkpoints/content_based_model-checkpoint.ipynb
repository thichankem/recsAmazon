# HỆ GỢI Ý THEO NỘI DUNG (CONTENT-BASED FILTERING)

Notebook này xây dựng mô hình gợi ý theo nội dung cho từng người dùng (User Profile) dựa trên đặc trưng sản phẩm (Item Profiles) được trích xuất bằng TF-IDF từ mô tả sản phẩm. Quy trình thực hiện gồm:
1. **Chuẩn bị dữ liệu**: Đọc dữ liệu từ file CSV đánh giá và thông tin sản phẩm.
2. **Xây dựng Item Profiles**: Vector hóa thông tin sản phẩm bằng TF-IDF.
3. **Xây dựng User Profiles (Mô hình người dùng)**: Đại diện sở thích của người dùng bằng cách tổng hợp các Item Profiles của sản phẩm họ đã từng đánh giá (có nhân với trọng số rating đã trừ đi trung bình).
4. **Đánh giá mô hình**: Tính toán dự đoán và đo lường sai số RMSE trên tập kiểm thử (Test Set).

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
import scipy.sparse as sp

# 1. Đọc dữ liệu đánh giá và thông tin sản phẩm
print('Đang tải dữ liệu...')
df_reviews = pd.read_csv('All_Beauty.csv')
df_meta = pd.read_csv('meta_All_Beauty.csv')

# Chỉ lấy các cột cần thiết cho mô hình và loại bỏ dòng trống
df_reviews = df_reviews[['user_id', 'parent_asin', 'rating']].dropna()
df_reviews.columns = ['userId', 'productId', 'score']

print(f'Số lượng review ban đầu: {len(df_reviews)}')
print(f'Số lượng sản phẩm trong metadata: {len(df_meta)}')
```

```python
# Gộp các cột văn bản mô tả sản phẩm
df_meta_clean = df_meta[['parent_asin', 'title', 'features', 'description']].dropna(subset=['parent_asin'])
df_meta_clean.columns = ['productId', 'title', 'features', 'description']
df_meta_clean['content'] = (
    df_meta_clean['title'].fillna('') + ' ' + 
    df_meta_clean['features'].fillna('') + ' ' + 
    df_meta_clean['description'].fillna('')
)

# Chỉ giữ các sản phẩm có mô tả văn bản thực tế
df_meta_clean = df_meta_clean[df_meta_clean['content'].str.strip() != '']

# Lọc reviews chỉ cho sản phẩm có thông tin mô tả
df_reviews = df_reviews[df_reviews['productId'].isin(df_meta_clean['productId'])]

# Lọc người dùng có ít nhất 5 đánh giá và sản phẩm có ít nhất 10 đánh giá
user_counts = df_reviews['userId'].value_counts()
product_counts = df_reviews['productId'].value_counts()

df_filtered = df_reviews[
    df_reviews['userId'].isin(user_counts[user_counts >= 5].index) &
    df_reviews['productId'].isin(product_counts[product_counts >= 10].index)
]

print(f'Dữ liệu sau khi lọc - Số review: {len(df_filtered)}')
print(f'Số users duy nhất: {df_filtered["userId"].nunique()} | Số products duy nhất: {df_filtered["productId"].nunique()}')
```

```python
# Lấy danh sách sản phẩm duy nhất trong tập dữ liệu đã lọc
unique_products = list(df_filtered['productId'].unique())
df_meta_filtered = df_meta_clean[df_meta_clean['productId'].isin(unique_products)].drop_duplicates('productId')

# Ánh xạ ID sản phẩm sang index để định chỉ mục ma trận
product2idx = {p: i for i, p in enumerate(unique_products)}
df_meta_filtered['idx'] = df_meta_filtered['productId'].map(product2idx)
df_meta_filtered = df_meta_filtered.sort_values('idx')

# Xây dựng đặc trưng văn bản cho từng sản phẩm dùng TF-IDF (Item Profiles)
tfidf = TfidfVectorizer(max_features=2000, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_meta_filtered['content'])
print(f'Kích thước ma trận đặc trưng sản phẩm (Item Profiles): {tfidf_matrix.shape}')
```

```python
# Chia tập Train/Test theo tỉ lệ 80/20
train_df, test_df = train_test_split(df_filtered, test_size=0.2, random_state=42)
print(f'Tập huấn luyện (Train Set): {len(train_df)} | Tập kiểm thử (Test Set): {len(test_df)}')
```

```python
# Tính giá trị trung bình (mean) và độ lệch chuẩn (std) của điểm đánh giá cho mỗi user trên tập Train
user_mean = train_df.groupby('userId')['score'].mean().to_dict()
user_std = train_df.groupby('userId')['score'].std().fillna(1.0).to_dict()

unique_users = list(df_filtered['userId'].unique())
user2idx = {u: i for i, u in enumerate(unique_users)}

# Xây dựng ma trận rating thưa W đã được chuẩn hóa (Centering)
row_indices = []
col_indices = []
data_values = []

for _, row in train_df.iterrows():
    u = row['userId']
    p = row['productId']
    score = row['score']
    
    if u in user2idx and p in product2idx:
        u_idx = user2idx[u]
        p_idx = product2idx[p]
        row_indices.append(u_idx)
        col_indices.append(p_idx)
        # Centering: Trừ đi rating trung bình của chính user đó
        data_values.append(score - user_mean.get(u, 3.0))

W = sp.csr_matrix((data_values, (row_indices, col_indices)), shape=(len(unique_users), len(unique_products)))

# Tính toán ma trận User Profiles bằng tích ma trận: User Profiles = W * Item Profiles
print('Đang tìm mô hình (User Profiles) cho từng người dùng...')
user_profiles = W.dot(tfidf_matrix)
print(f'Kích thước ma trận mô hình người dùng (User Profiles): {user_profiles.shape}')
```

```python
# Chuẩn hóa vector hàng của User Profiles và Item Profiles để tính Cosine Similarity nhanh hơn bằng tích vô hướng
from sklearn.preprocessing import normalize
user_profiles_norm = normalize(user_profiles, norm='l2', axis=1)
tfidf_matrix_norm = normalize(tfidf_matrix, norm='l2', axis=1)

sq_errors = []
predictions = []
actuals = []

print('Đang dự đoán điểm và đánh giá mô hình trên tập kiểm thử (Test Set)...')
for _, row in test_df.iterrows():
    u = row['userId']
    p = row['productId']
    actual = row['score']
    
    if u in user2idx and p in product2idx:
        u_idx = user2idx[u]
        p_idx = product2idx[p]
        
        # Lấy vector đại diện của user và product đã chuẩn hóa
        u_vec = user_profiles_norm[u_idx]
        p_vec = tfidf_matrix_norm[p_idx]
        
        # Cosine Similarity là tích vô hướng
        cos_sim = u_vec.dot(p_vec.T)[0, 0]
        
        # Dự đoán rating: mean của user + cos_sim * std của user * hệ số tỷ lệ
        mean_u = user_mean.get(u, 3.0)
        std_u = user_std.get(u, 1.0)
        
        pred = mean_u + cos_sim * std_u * 1.5
        pred = np.clip(pred, 1.0, 5.0) # Đưa về giới hạn rating thực tế [1, 5]
        
        sq_errors.append((actual - pred) ** 2)
        predictions.append(pred)
        actuals.append(actual)

rmse_score = np.sqrt(np.mean(sq_errors))
print(f'==> SAI SỐ RMSE CỦA MÔ HÌNH CONTENT-BASED: {rmse_score:.4f}')

print('\nMột số ví dụ dự đoán thực tế vs Điểm số thật:')
for act, pre in list(zip(actuals, predictions))[:10]:
    print(f'  Thực tế: {act:.1f} sao | Dự đoán: {pre:.2f} sao')
```

```python
def recommend_content_based(user_id, top_n=5):
    if user_id not in user2idx:
        print('User mới hoặc không có dữ liệu huấn luyện!')
        return []
    
    u_idx = user2idx[user_id]
    u_vec = user_profiles_norm[u_idx]
    
    # Tính toán similarity của user này với TOÀN BỘ sản phẩm
    scores = tfidf_matrix_norm.dot(u_vec.T).toarray().flatten()
    
    # Lấy các sản phẩm mà user CHƯA từng đánh giá trong tập train
    rated_products_indices = W[u_idx].nonzero()[1]
    
    recommendations = []
    for p_idx, product_id in enumerate(unique_products):
        if p_idx not in rated_products_indices:
            recommendations.append((product_id, scores[p_idx]))
            
    # Sắp xếp theo score tương quan giảm dần
    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations[:top_n]

# Gợi ý cho một user ngẫu nhiên
sample_user_id = df_filtered['userId'].iloc[0]
recs = recommend_content_based(sample_user_id, top_n=5)
print(f'Gợi ý sản phẩm cho người dùng: {sample_user_id}')
for pid, sim_score in recs:
    # Tìm title của sản phẩm trong metadata
    title = df_meta_clean[df_meta_clean['productId'] == pid]['title'].iloc[0] if pid in df_meta_clean['productId'].values else 'No Title'
    print(f'  Mã: {pid} | Tên: {title[:70]}... | Similarity: {sim_score:.4f}')
```
