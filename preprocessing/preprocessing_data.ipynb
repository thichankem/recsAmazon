## 1. User Reviews (Đánh giá từ người dùng)

| Tên trường (Field) | Kiểu dữ liệu (Type) | Mô tả (Explanation) |
| :--- | :--- | :--- |
| **`rating`** | `float` | Điểm đánh giá sản phẩm (từ `1.0` đến `5.0`). |
| **`title`** | `str` | Tiêu đề của bài đánh giá. |
| **`text`** | `str` | Nội dung chi tiết của bài đánh giá. |
| **`images`** | `list` | Danh sách hình ảnh do người dùng đăng tải sau khi nhận hàng. Mỗi ảnh gồm nhiều kích thước khác nhau (`small_image_url`, `medium_image_url`, `large_image_url`). |
| **`asin`** | `str` | ID định danh duy nhất của phiên bản sản phẩm cụ thể. |
| **`parent_asin`** | `str` | ID gốc (Parent ID) của sản phẩm. *Lưu ý: Các sản phẩm khác màu, kiểu dáng, kích thước thường dùng chung một parent_asin. Hãy dùng ID này để liên kết với bảng Item Metadata.* |
| **`user_id`** | `str` | ID của người viết đánh giá. |
| **`timestamp`** | `int` | Thời gian đăng đánh giá (định dạng Unix timestamp). |
| **`verified_purchase`** | `bool` | Xác thực người dùng đã mua hàng thực tế hay chưa (`True`/`False`). |
| **`helpful_vote`** | `int` | Số lượt bình chọn "hữu ích" (helpful) từ những người dùng khác cho đánh giá này. |

---

## 2. Item Metadata (Thông tin chi tiết sản phẩm)

| Tên trường (Field) | Kiểu dữ liệu (Type) | Mô tả (Explanation) |
| :--- | :--- | :--- |
| **`main_category`** | `str` | Danh mục chính (ngành hàng/domain) của sản phẩm. |
| **`title`** | `str` | Tên đầy đủ của sản phẩm. |
| **`average_rating`** | `float` | Điểm đánh giá trung bình hiển thị trên trang sản phẩm. |
| **`rating_number`** | `int` | Tổng số lượt đánh giá của sản phẩm này. |
| **`features`** | `list` | Các tính năng nổi bật của sản phẩm (dưới dạng danh sách gạch đầu dòng). |
| **`description`** | `list` | Phần mô tả chi tiết về sản phẩm. |
| **`price`** | `float` | Giá bán của sản phẩm bằng USD (tại thời điểm thu thập dữ liệu). |
| **`images`** | `list` | Danh sách hình ảnh của sản phẩm. Mỗi ảnh có các kích thước khác nhau (`thumb`, `large`, `hi_res`). Trường `variant` biểu thị vị trí hiển thị của ảnh. |
| **`videos`** | `list` | Danh sách video liên quan đến sản phẩm (bao gồm tiêu đề và đường dẫn URL). |
| **`store`** | `str` | Tên của cửa hàng/thương hiệu bán sản phẩm. |
| **`categories`** | `list` | Danh mục phân cấp của sản phẩm (từ lớn đến nhỏ). |
| **`details`** | `dict` | Thông tin kỹ thuật chi tiết (chất liệu, thương hiệu, kích thước, v.v.) dưới dạng khóa-giá trị (key-value). |
| **`parent_asin`** | `str` | ID gốc (Parent ID) của sản phẩm (dùng để liên kết với bảng Reviews). |
| **`bought_together`** | `list` | Danh sách các gói sản phẩm được đề xuất mua cùng nhau từ website. |

```python
import pandas as pd

reviews_file = '../data/Cell_Phones_and_Accessories.jsonl'
meta_file = '../data/meta_Cell_Phones_and_Accessories.jsonl'
```

```python
first_chunk = True

meta_chunks = pd.read_json(meta_file, lines=True, chunksize=100000)

processed_chunks = []

for m_chunk in meta_chunks:
    m_chunk.drop(columns=['images', 'details', 'bought_together', 'videos', 'price', 'average_rating', 'rating_number', 'store', 'subtitle', 'author'], inplace=True)
    processed_chunks.append(m_chunk)

df_meta = pd.concat(processed_chunks, ignore_index=True)
df_meta
```

```python
first_chunk = True

review_chunks = pd.read_json(reviews_file, lines=True, chunksize=10000)

processed_chunks = []

for r_chunk in review_chunks:
    r_chunk.drop(columns=['images', 'verified_purchase', 'text', 'title', 'helpful_vote'], inplace=True)
    processed_chunks.append(r_chunk)
    if len(processed_chunks) >= 100:
        break

df_review = pd.concat(processed_chunks, ignore_index=True)
df_review
```

```python
df_joined = pd.merge(df_review, df_meta, on='parent_asin', how='left')

df_joined['content'] = df_joined['features'].fillna('') + df_joined['description'].fillna('')

df_joined.drop(columns=['categories', 'features', 'description'], inplace=True)

df_joined
```

```python
df_joined.to_csv('preprocessed_data.csv',  index=False,  chunksize=50000)

print("Đã lưu file thành công!")
```
